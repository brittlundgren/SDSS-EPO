## III. Querying the Sloan Digital Sky Survey's database (SDSS DR16)
The next code block searches the SDSS Data Release 16 database via the CasJobs REST API. The query completes quickly, so it uses CasJobs quick mode.
CasJobs also has an asynchronous mode, which will submit job to a queue and will store the results in a table in your MyDB. If your results are very large, it will store the results in MyScratchDB instead.

Run the code block below to query DR16. Try changing some of the query parameters at the top of the code block below to see how the output changes.

Documentation on the SciServer Python libraries can be found on the SciServer Python documentation site at:
http://www.sciserver.org/docs/sciscript-python/

An extensive tutorial on how to query the SDSS database can be found here:
http://cas.sdss.org/dr16/en/help/howto/search/searchhowtohome.aspx

The paper describing the SDSS Data Release 16 is available at the following link: https://arxiv.org/abs/1912.02905

```python
# Find objects in the Sloan Digital Sky Survey's Data Release 16.
#
# Query the Sloan Digital Sky Serveys' Data Release 16.
# For the database schema and documentation see http://skyserver.sdss.org/dr16
#
# This query finds "a 4x4 grid of nice-looking galaxies": 
#   galaxies in the SDSS database that have a spectrum 
#   and have a size (petror90_r) larger than 10 arcsec.
# 
# First, store the query in an object called "query"
query="""
SELECT TOP 16 p.objId,p.ra,p.dec,p.petror90_r, p.g, p.r
  FROM galaxy AS p
   JOIN SpecObj AS s ON s.bestobjid = p.objid
WHERE p.u BETWEEN 0 AND 19.6
  AND p.g BETWEEN 0 AND 17  AND p.petror90_r > 10
"""
gals = CasJobs.executeQuery(query, "dr16")
gals = gals.set_index('objId')
gals
```

### Generating a galaxy color-magnitude diagram

```python
# enter query
query="""
SELECT TOP 1000 p.objId,p.ra,p.dec,p.petror90_r, p.u,p.g,p.r,p.i,p.z
  FROM galaxy AS p
   JOIN SpecObj AS s ON s.bestobjid = p.objid
WHERE p.u BETWEEN 0 AND 19.6
  AND p.g BETWEEN 0 AND 17
  AND p.petror90_r > 10
"""
# send query to CasJobs
gals = CasJobs.executeQuery(query, "dr16")
# show results
gals
```

```python
# make scatterplot
gal_color = gals['g']-gals['r']
gal_mag = gals['g']
plt.scatter(gal_color, gal_mag, alpha=0.5)
plt.xlabel('g-r')
plt.ylabel('g')
plt.show() 
```

A scatter plot with a lot of overlapping data points can be challenging to understand. The "alpha" parameter in the "scatter" function allows you to modify the transparency of markers on the plot. You may want to try changing this value within the range 0-1 to see the difference it makes.  If even a very low value for the "alpha" parameter does not clarify densely clustered points, a color-coded density plot (shown below) is just one of various outher ways you can make a crowded plot more understandable.

```python

nbins=40
plt.hist2d(gal_color, gal_mag, bins=nbins)
plt.xlabel('g-r')
plt.ylabel('g')
plt.colorbar()
plt.show()
```


## VI. Retrieve thumbnail cutouts of galaxies and show them on screen
SkyServer has a service that will produce a color image cutout of certain dimensions around a specified position, displayed as a JPG thumbnail.

The code below iterates through each galaxy in your results and calls the image cutout generator for each galaxy. The scale of the image depends on the Petrosian radius of the galaxy.

```python
# set thumbnail parameters
width=200           # image width
height=200          # height
pixelsize=0.396     # image scale
plt.figure(figsize=(15, 15))   # display in a 4x4 grid
subPlotNum = 1

i = 0
nGalaxies = len(gals)
for index,gal in gals.iterrows():           # iterate through rows in the DataFrame
    i = i + 1
    if i<17:
        print('Getting image '+str(i)+' of '+str(nGalaxies)+'...')
        if (i == nGalaxies):
            print('Plotting images...')
        scale=2*gal['petror90_r']/pixelsize/width
        img= SciServer.SkyServer.getJpegImgCutout(ra=gal['ra'], dec=gal['dec'], width=width, height=height, scale=scale,dataRelease='DR13')
        plt.subplot(4,4,subPlotNum)
        subPlotNum += 1
        plt.imshow(img)                               # show images in grid
        plt.title(index)                              # show the object identifier (objId) above the image.
```
